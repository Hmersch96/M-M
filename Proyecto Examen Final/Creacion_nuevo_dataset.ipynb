```python
def indicador_notaFinal(nota):
    if nota==1.0:
        salida=[1,0,0,0,0]
    elif nota==2.0:
        salida=[0,1,0,0,0]
    elif nota==3.0:
        salida=[0,0,1,0,0]
    elif nota==4.0:
        salida=[0,0,0,1,0]
    elif nota==5.0:
        salida=[0,0,0,0,1]
    else:
        salida=[0,0,0,0,0]
    return salida

for estudiante in Idlists:
    #Aqui se almacena todos los datos disponibles de datosConcatenados de los alumnos que pertenecen a la lista Idlists
    datosEstudiante=datosConcatenados[datosConcatenados['id_anony']==estudiante]
    #Aqui se obtienen TODAS las materias que cursan este conjunto de alumnos
    #materias=np.unique(datosEstudiante.Asignatura.values)
    registro['id_anony']=estudiante
              
    k=0
    
    for materia in materias:
        k=k+1
        #Aqui se almacena todas las ocasiones en la que el estudiante curso la materia
        EstudianteMateria=datosEstudiante[datosEstudiante['Asignatura']==materia]
        
        #j indica la cantidad de intentos que le tomo al estudiante pasar la materia
        j=1
        desercion=0
        for index, row in  EstudianteMateria.iterrows():
            #Aqui se extraen los datos que seran importados en el nuevo dataset
            
            #Aqui esta el codigo que trato de hacer para obtener los resultados de cada caso

#             registro['Cod.Car.Sec']=row['Cod.Car.Sec']
#             registro['Cod.Car.Sec Codificado']=row['Cod.Car.Sec Codificado']
#             registro['Materia_'+str(k)]=row['Asignatura']
#             registro['Materia_'+str(k)+' codificada']=row['Asignatura Codificada']
#             registro['Anho_F_'+str(k)+' Intento '+str(j)]=row['Anho']
#             registro['ciclo_F_'+str(k)+' Intento '+str(j)]=row['Ciclo']     
#             registro['P1_F_'+str(k)+' Intento '+str(j)]=row['1P']
#             registro['P2_F_'+str(k)+' Intento '+str(j)]=row['2P']
#             registro['T_F_'+str(k)+' Intento '+str(j)]=row['Taller']
#             registro['NotaFinal_'+str(k)+' Intento '+str(j)]=row['Nota_Final']
            
#             registro['Materia_'+str(k)+' Nota=1'],registro['Materia_'+str(k)+' Nota=2'],registro['Materia_'+str(k)+' Nota=3'],registro['Materia_'+str(k)+' Nota=4'],registro['Materia_'+str(k)+' Nota=5']=indicador_notaFinal(row['Nota_Final'])

#             if row['Nota_Final']==0.0:
#                 desercion=desercion+1
#                 registro['Deserto_Materia'+str(k)+' intento'+str(j)]=1
#             else:
#                 registro['Deserto_Materia'+str(k)+' intento'+str(j)]=0
#             j=j+1

            #Esto se usa si no me funciona lo de ver CADA vez que curso
            registro['Cod.Car.Sec']=row['Cod.Car.Sec']
            registro['Cod.Car.Sec Codificado']=row['Cod.Car.Sec Codificado']
            registro['Materia_'+str(k)]=row['Asignatura']
            registro['Materia_'+str(k)+' codificada']=row['Asignatura Codificada']
            registro['Anho_F_'+str(k)]=row['Anho']
            registro['ciclo_F_'+str(k)]=row['Ciclo']  
            
            
            registro['P1_F_'+str(k)]=row['1P']
            registro['P2_F_'+str(k)]=row['2P']
            registro['T_F_'+str(k)]=row['Taller']
            registro['NotaFinal_'+str(k)]=row['Nota_Final']
            
            
            registro['Materia_'+str(k)+' Nota=1'],registro['Materia_'+str(k)+' Nota=2'],registro['Materia_'+str(k)+' Nota=3'],registro['Materia_'+str(k)+' Nota=4'],registro['Materia_'+str(k)+' Nota=5']=indicador_notaFinal(row['Nota_Final'])
            if row['Nota_Final']==0.0:
                desercion=desercion+1
                
        registro['Abandono la Materia '+str(k)]=desercion    
        #Aqui se cuenta la cantidad de ocasiones en la que el estudiante recurso la materia
        
        
        
    Historico=pd.concat([Historico, registro], axis=0)


```
