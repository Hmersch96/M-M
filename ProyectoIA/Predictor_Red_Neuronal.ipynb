```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Importamos el dataset
df = pd.read_csv("Resumido.csv")

#Realizamos las ultimas modificaciones en los datos
del df['Unnamed: 0']

df['Aprobo' ] =df['Aprobado'].replace(['N','S'],['0','1'])
df['Derecho a Recuperatorio' ] =df['Derecho a Recuperatorio'].replace([False,True],['0','1'])
df['Derecho a Final' ] =df['Derecho a Final'].replace([False,True],['0','1'])
```

```python
data=df[['1P','2P','Taller']].values
labels=df[['Aprobo']].values
#Como labels figura como string, es necesari cambiarlos a valores enteros
labels = labels.astype(np.int)
labels=np.ravel(labels)
print(data.shape)
print(labels.shape)
```

```python
labels
```

```python
#Aqui se separan los datos de entrenamiento y testeo
data_train, data_test, labels_train, labels_test = train_test_split(data, labels)
```

```python
#Creamos el modelo
model = Sequential()
model.add(Dense(5, input_shape=(3,),activation='sigmoid' , name = "Neurona_de_Entrada"))
model.add(Dense(10, activation='sigmoid' , name = "Neurona_1"))
model.add(Dense(10, activation='sigmoid' , name = "Neurona_2"))

model.add(Dense(1, activation='sigmoid' , name = "Neurona_de_Salida"))

#aqui se habla del metodo compile
#https://keras.io/api/layers/activations/

#Aqui se habla del metodo Dense
#https://keras.io/api/layers/core_layers/dense/

#Aqui esta una lista de las activaciones
#https://www.tensorflow.org/api_docs/python/tf/keras/losses
```

```python
#Configuramos el modelo para emplear regresion logistica
model.compile(loss='binary_crossentropy', optimizer='Adam')

#Para saber como funciona compile
#https://keras.io/api/models/model_training_apis/

#Para saber las opciones en loss
#https://www.tensorflow.org/api_docs/python/tf/keras/losses
#https://keras.io/api/losses/

#Para saber las opciones en optimizers
#https://www.tensorflow.org/api_docs/python/tf/keras/optimizers
```

```python
model.summary()
```

```python
#Se entrena al modelo
history = model.fit(data_train, labels_train, epochs=1000, batch_size=10, verbose=1, validation_split=0.2)
#epochs: iteraciones al recorrer los datos
#batch_size: Tamanho de lotes de datos

model.summary()
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
# Evaluar
print(model.evaluate(data, labels))
print(model.metrics_names)
```

```python
#Ahora veerificamos la precision del modelo
from sklearn.metrics import confusion_matrix, classification_report 
from sklearn.metrics import accuracy_score
predictions = model.predict(data_test)
predict_label = predictions.reshape(-1).round()

print('Cantidad de predicciones = {} \n'.format(predict_label.shape))

cf_matrix = confusion_matrix(labels_test, predict_label)
print(cf_matrix)

exactitud=accuracy_score(labels_test, predict_label)*100
print('Exactitud del modelo: ',exactitud,' %')
```

```python
sns.set_theme(color_codes=True)
sns.heatmap(cf_matrix, annot=True)
```
