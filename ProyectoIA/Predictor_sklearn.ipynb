```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Importamos el dataset
df = pd.read_csv("Resumido.csv")

#Realizamos las ultimas modificaciones en los datos
del df['Unnamed: 0']

df['Aprobo' ] =df['Aprobado'].replace(['N','S'],['0','1'])
df['Derecho a Recuperatorio' ] =df['Derecho a Recuperatorio'].replace([False,True],['0','1'])
df['Derecho a Final' ] =df['Derecho a Final'].replace([False,True],['0','1'])
```

```python
display(df)
```

<h1>Extraccion y verificacion de datos</h1>

```python
#Extraemos la(s) variables de salida
y=df[['Aprobo']]

#Extraemos las variables de entrada
X=df[['1P','2P','Taller']]
```

```python
#Se revisa la cantidad de variables en cada vector. Este paso se realiza para revisar si hay valores faltantes
display(X.shape)
display(y.shape)

#En caso de que hayan datos faltantes, es necesario sacarlos de los datos para que no interfieran con la red neuronal
```

```python
#Convertimos los datos a Numpy
X=np.array(X)
y=np.array(y)

#Esta funcion altera la dimensionn de y para que no genere errores en el modelo
y=np.ravel(y)
```

<h1>Creamos el Modelo</h1>

```python
logmod=LogisticRegression()
```

```python
logmod.fit(X,y)
```

<h1>Comprobacion de Precision</h1>

```python
print('Precision: ',logmod.score(X,y)*100,'%')
```
